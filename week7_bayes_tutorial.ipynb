Step 1: Prior Probabilities

P(playing = yes) = 9 / 14 = 0.642

P(playing = no) = 5 / 14 = 0.357

Step 2: Conditional Probabilities

For playing = yes

P(x = sunny | yes) = 2 / 9 = 0.222
P(x = hot | yes) = 2 / 9 = 0.222
P(x = high | yes) = 3 / 9 = 0.333
P(x = false | yes) = 6 / 9 = 0.667

For playing = no

P(x = sunny | no) = 3 / 5 = 0.6
P(x = hot | no) = 2 / 5 = 0.4
P(x = high | no) = 4 / 5 = 0.8
P(x = false | no) = 2 / 5 = 0.4

Step 3: Naive Bayes Calculation

For playing = yes

P(playing = yes | x)
= 0.222 × 0.222 × 0.333 × 0.667 × 0.642
≈ 0.0068

For playing = no

P(playing = no | x)
= 0.6 × 0.4 × 0.8 × 0.4 × 0.357
≈ 0.0274

Conclusion

Since:
P(playing = no | x) > P(playing = yes | x)

Final Prediction: Playing = No
